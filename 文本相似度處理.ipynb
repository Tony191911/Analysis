```python
def split_text(text, stop_words_path):
    text = re.sub(r'[^a-zA-Z\u4e00-\u9fff]', '', text, re.I|re.A)
    text = re.sub(r'\n','',text)
    tokens = list(jieba.cut(text,cut_all=False))
    result = [t for t in tokens if t.strip()]
    tokenized_text = ' '.join(tokens)
    with open(stop_words_path, 'r',encoding='utf-8') as f:
        st = f.read()
        st_list = st.splitlines()
    word_list = tokenized_text.split()
    
    cleaned_words = [word for word in word_list if word not in st_list]
    return cleaned_words

path = 'C:\\Users\\fento\\Downloads\\data_cleaned\\gamesdata_280_2.xlsx'
stop_words_path = 'C:\\Users\\fento\\Downloads\\gameID_reviews_cleaned\\stop_words_3.txt'
df = pd.read_excel(path,index_col=0)

df['names'] = df['名字'].apply(lambda x: split_text(x,stop_words_path))
df['tags'] = df['標籤'].apply(lambda x: x.split('\n'))
df['description'] = df['描述'].apply(lambda x: split_text(x,stop_words_path))
df['reviews'] = df['評論'].apply(lambda x: split_text(x,stop_words_path))

output_path = 'C:\\Users\\fento\\Downloads\\data_cleaned\\gamesdata_280_2.xlsx'
df.to_excel(output_path, index=False) 
df[['names' , 'tags' , 'description', 'reviews']].head()
```

```python
corpus = df['names'].tolist() + df['tags'].tolist() + df['description'].tolist() + df['reviews'].tolist()

# Train a Word2Vec model
word2vec_model = Word2Vec(sentences=corpus, vector_size=100, window=5, min_count=1, workers=4)
word2vec_model.train(corpus, total_examples=len(corpus), epochs=10)

# Test the Word2Vec model
similar_words = word2vec_model.wv.most_similar('射擊', topn=15)
word2vec_model.save("C:\\Users\\fento\\Downloads\\data_cleaned\\word2vec.model")
similar_words
```

```python
# Importing necessary modules
import pandas as pd
import jieba
from gensim.models import Word2Vec

# Reading the data
games_data = pd.read_excel("C:\\Users\\fento\\Downloads\\data_cleaned\\gamesdata_280_2.xlsx")
```

```python
# Load the pre-trained Word2Vec model
word_vec_model = Word2Vec.load(r"C:\Users\fento\Downloads\model\word2vec_2.model")

def text_to_vector(text, model):
    vector_list = [model.wv[word] for word in text if word in model.wv]
    if len(vector_list) == 0:
        return np.zeros(model.vector_size)
    return np.mean(vector_list, axis=0)

games_data['tags_vector'] = games_data['tags'].apply(lambda x: text_to_vector(x, word2vec_model))
games_data['description_vector'] = games_data['description'].apply(lambda x: text_to_vector(x, word2vec_model))
games_data['reviews_vector'] = games_data['reviews'].apply(lambda x: text_to_vector(x, word2vec_model))

output_path = 'C:\\Users\\fento\\Downloads\\data_cleaned\\gamesdata_280_2.xlsx'
games_data.to_excel(output_path, index=False) 
games_data[['tags', 'tags_vector', 'description', 'description_vector', 'reviews', 'reviews_vector']].head()
```

```python
def text_to_vector(text, model):
    vector_list = [model.wv[word] for word in text if word in model.wv]
    if len(vector_list) == 0:
        return np.zeros(model.vector_size)
    return np.mean(vector_list, axis=0)
def split_text(user_input):
    text_generater = jieba.cut(user_input,cut_all=False)
    result = ' '.join(text_generater)
    return result

def convert_str_to_array(s):
    return np.array(s[1:-1].split(), dtype=float)
games_data[['tags_vector','description_vector','reviews_vector']].head()
```

```python
from sklearn.metrics.pairwise import euclidean_distances

def recommend_games(user_input, games_data, top_n=10):
    processed_input = split_text(user_input)
    input_vector = text_to_vector(processed_input, word2vec_model)
    input_vector = input_vector.reshape(1, -1)

    tags_similarity = euclidean_distances(input_vector, np.stack(games_data['tags_vector'].apply(convert_str_to_array)))
    description_similarity = euclidean_distances(input_vector, np.stack(games_data['description_vector'].apply(convert_str_to_array)))
    reviews_similarity = euclidean_distances(input_vector, np.stack(games_data['reviews_vector'].apply(convert_str_to_array)))
    sentiment_score = games_data['Sentiment_Score']
    sentiment_score_array = sentiment_score.values.reshape(-1, 1)
    tags_weight = 0.4
    description_weight = 0.3
    reviews_weight = 0.2
    sentiment_weight = 0.1

    avg_similarity = (tags_similarity * tags_weight + description_similarity * description_weight + 
    reviews_similarity * reviews_weight + sentiment_score_array * sentiment_weight) / (tags_weight + description_weight + reviews_weight + sentiment_weight)
    
    # Get the top N most similar games
    top_n_indices = avg_similarity[0].argsort()[-top_n:][::-1]
    top_n_games = games_data.iloc[top_n_indices]
    
    return top_n_games[['名字', '描述']]

path = 'C:\\Users\\fento\\Downloads\\data_cleaned\\gamesdata_280_2.xlsx'
stop_words_path = 'C:\\Users\\fento\\Downloads\\gameID_reviews_cleaned\\stop_words_3.txt'
games_data = pd.read_excel(path,index_col=0)
# Test the recommendation system with a sample user input
user_input = "第一人稱射擊"
recommend_games(user_input, games_data)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_games(user_input, games_data, top_n=10):
    # 使用者輸入(分詞)
    processed_input = split_text(user_input)
    # 將處理後的輸入轉換成向量。
    input_vector = text_to_vector(processed_input, word2vec_model)
    input_vector = input_vector.reshape(1, -1)
    
    # 計算使用者輸入與所有遊戲之間標籤、描述、評論的餘弦相似度。
    tags_similarity = cosine_similarity(input_vector, np.stack(games_data['tags_vector'].apply(convert_str_to_array)))
    description_similarity = cosine_similarity(input_vector, np.stack(games_data['description_vector'].apply(convert_str_to_array)))
    reviews_similarity = cosine_similarity(input_vector, np.stack(games_data['reviews_vector'].apply(convert_str_to_array)))
    sentiment_score = games_data['Sentiment_Score']
    sentiment_score_array = sentiment_score.values.reshape(-1, 1)
    tags_weight = 0.3
    description_weight = 0.4
    reviews_weight = 0.2
    sentiment_weight = 0.1

    avg_similarity = (tags_similarity * tags_weight + description_similarity * description_weight + 
    reviews_similarity * reviews_weight + sentiment_score_array * sentiment_weight) / 
    (tags_weight + description_weight + reviews_weight + sentiment_weight)
    
    # 取得前N個最相似的遊戲
    top_n_indices = avg_similarity[0].argsort()[-top_n:][::-1]
    top_n_games = games_data.iloc[top_n_indices]
    
    return top_n_games[['名字', '描述', '價格']]

path = 'C:\\Users\\fento\\Downloads\\data_cleaned\\gamesdata_280_2.xlsx'
stop_words_path = 'C:\\Users\\fento\\Downloads\\gameID_reviews_cleaned\\stop_words_3.txt'
games_data = pd.read_excel(path,index_col=0)
# 使用者輸入測試
user_input = "第一人稱射擊"
recommend_games(user_input, games_data)
```
